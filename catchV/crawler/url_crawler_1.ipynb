```python
def pornhub(input1):
    if input1 == 'asian':
        cat = "video?c=1"
    url = 'https://www.pornhub.com/'+ cat
    link_list = []
    time_list = []
    cnt = 1
    next_page= [1]
    while next_page:   #(cnt != 3)
        driver.get(url)
        wait = WebDriverWait(driver, 10)
        element = wait.until(EC.element_to_be_clickable((By.ID, 'js-networkBar')))
        hrefs = driver.find_elements(By.XPATH, '//div[@class="phimage"]/a')
        srcs = driver.find_elements(By.XPATH, '//div[@class="phimage"]/a/img')
        for h, s in zip(hrefs, srcs):
            link_list.append('https://www.pornhub.com'+h.get_dom_attribute('href'))
            time_list.append(s.get_dom_attribute('data-thumb_url').split('/')[4])
        cnt += 1
        url = 'https://www.pornhub.com/'+cat+'&page='+"{}".format(cnt)
        next_page = driver.find_elements(By.XPATH, '//a[@class="orangeButton"]')
    df = pd.DataFrame((zip(link_list, time_list)), columns = ['link', 'upload_date'])
    return df
```

```python
def gotporn(cat):

    url = "https://www.gotporn.com/categories/"+cat
    link_list = []
    time_list = []
    title_list = []
    cnt = 1
    next_page = [1]
    while next_page:   # next_page
        driver.get(url)
        wait = WebDriverWait(driver, 10)
        element = wait.until(EC.element_to_be_clickable((By.XPATH, '//*[@id="wrapper"]/header/div[1]/div[1]/a/img')))
        #btnAdult =  driver.find_elements(By.XPATH, "/html/body/section/div/div[2]/div")
        #if btnAdult:
            #btnAdult[0].click()
        hrefs = driver.find_elements(By.XPATH, '//li[@class="video-item  poptrigger "]/a')
        srcs = driver.find_elements(By.XPATH, '//li[@class="video-item  poptrigger "]/a/span/span[2]/img')
        for h, s in zip(hrefs, srcs):
            '''
            try:
                title_list.append(h.get_dom_attribute('data-title'))
            except:
                title_list.append('None')
            '''
            link_list.append(h.get_dom_attribute('href'))
            
            time_list.append("".join(s.get_dom_attribute('data-default-src').split('/')[3:5]))
        
            
        cnt += 1
        url = 'https://www.gotporn.com/'+"categories/"+cat+'?page='+"{}".format(cnt)
        next_page = driver.find_elements(By.XPATH, '//a[@class="btn btn-secondary paginate-show-more"]')
    df = pd.DataFrame((zip(link_list, time_list)), columns = ['link', 'upload_date']) # title_list
    return df
```

```python
def spankwire(cat):
    cat = cat
    url = 'https://www.redtube.com/redtube/'+cat
    link_list = []
    time_list = []
    title_list = []
    cnt = 1
    next_page = [1]
    while next_page:   #(cnt != 5)
        driver.get(url)
        wait = WebDriverWait(driver, 10)
        element = wait.until(EC.element_to_be_clickable((By.XPATH, '//*[@id="redtube_logo_image"]')))
        try:
            hrefs = driver.find_elements(By.XPATH, '//div[@class="video_block_wrapper js_mediaBookBounds "]/span/a')
            srcs = driver.find_elements(By.XPATH, '//div[@class="video_block_wrapper js_mediaBookBounds "]/span/a/picture/img')
            for h, s in zip(hrefs, srcs):
                try:
                    link_list.append('https://www.redtube.com'+h.get_dom_attribute('href'))
                    time_list.append("".join(s.get_dom_attribute('data-path').split('/')[4]))
                except:
                    pass
        except:
            pass
            
        cnt += 1
        url = 'https://www.redtube.com/'+"redtube/"+cat+'?page='+"{}".format(cnt)
        next_page = driver.find_elements(By.XPATH, '//a[@id="wp_navNext"]')
        
    df = pd.DataFrame((zip(link_list, time_list)), columns = ['link', 'upload_date']) # title_list
    return df
```

```python
def main(cat):
    df3 = spankwire(cat)
    df1 = pornhub(cat)
    df2 = gotporn(cat)
    
    all_df = pd.concat([df1, df2]) # df3
    all_df = all_df.drop_duplicates(['link'])
    return all_df
```
